# Example 7-1

Code to generate k-means examples

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
n_data = 1000
seed = 1
n_centers = 4
```

```python
# Generate random Gaussian blobs and run K-means
blobs, blob_labels = make_blobs(n_samples=n_data, n_features=2, 
                                centers=n_centers, random_state=seed)
clusters_blob = KMeans(n_clusters=n_centers, random_state=seed).fit_predict(blobs)
```

```python
# Generate data uniformly at random and run K-means
uniform = np.random.rand(n_data, 2)
clusters_uniform = KMeans(n_clusters=n_centers, random_state=seed).fit_predict(uniform)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 5))

ax1.scatter(blobs[:, 0], blobs[:, 1], c=blob_labels, edgecolors='k', cmap='gist_rainbow')
ax1.set_title("(a) Four randomly generated blobs", fontsize=14)
ax1.set_axis_off()

ax2.scatter(blobs[:, 0], blobs[:, 1], c=clusters_blob, edgecolors='k', cmap='gist_rainbow')
ax2.set_title("(b) Clusters found via K-means", fontsize=14)
ax2.set_axis_off()

ax3.scatter(uniform[:, 0], uniform[:, 1], edgecolors='k')
ax3.set_title("(c) 1000 randomly generated points", fontsize=14)
ax3.set_axis_off()

ax4.scatter(uniform[:, 0], uniform[:, 1], c=clusters_uniform, edgecolors='k', cmap='gist_rainbow')
ax4.set_title("(d) Clusters found via K-means", fontsize=14)
ax4.set_axis_off()
```

# Example 7-2

k-means on the Swiss roll

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import make_swiss_roll
```

```python
# Generate a noisy Swiss roll dataset
X, color = make_swiss_roll(n_samples=1500)
```

```python
# Approximate the data with 100 k-means clusters
clusters_swiss_roll = KMeans(n_clusters=100, random_state=seed).fit_predict(X)
```

```python
# Plot the dataset with k-means cluster IDs as the color
fig2 = plt.figure()
ax = fig2.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=clusters_swiss_roll, cmap='Spectral')
plt.show()
```
