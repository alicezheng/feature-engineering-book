# Example 8-2

Applying a simple Gaussian filter on an image

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage import data, color
from scipy import signal
```

```python
# First create X,Y meshgrids of size 5x5 on which we compute the Gaussian
ind = [-1., -0.5, 0., 0.5, 1.]
X, Y = np.meshgrid(ind, ind)
```

```python
X
```

```python
Y
```

```python
# G is the unnormalized Gaussian kernel where the value at (0, 0) is 1.0
G = np.exp(-(X*X + Y*Y) / 2)
G
```

```python
cat = color.rgb2gray(data.chelsea())
```

```python
blurred_cat = signal.convolve2d(cat, G, mode='valid')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, 
                               figsize=(10, 4), 
                               sharex=True, 
                               sharey=True)

ax1.axis('off')
ax1.imshow(cat, cmap=plt.cm.gray)
ax1.set_title('Input image')
ax1.set_adjustable('box')

ax2.axis('off')
ax2.imshow(blurred_cat, cmap=plt.cm.gray)
ax2.set_title('After convolving with a Gaussian filter')
ax2.set_adjustable('box')

plt.tight_layout()
plt.show()
```
